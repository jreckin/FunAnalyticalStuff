```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
ccm_stats = pd.read_csv('ccm_shot_chart_and_stats.csv')
```

```python
ccm_stats.dtypes
```

```python
# Drop non-numeric values 
ccm_stats_num = ccm_stats.drop(columns = {'GameId', 'Date', 'Opponent'})
```

```python
# Replace win and loss with 1 and 0 to make it numeric
ccm_stats_num = ccm_stats_num.replace(to_replace = ['Win','Loss'],value = ['1','0'])
```

```python
# Cast win and loss (1 and 0) now as ints instead of strings
ccm_stats_num['WinOrLoss'] = ccm_stats_num['WinOrLoss'].astype(str).astype(int)
```

```python
# Check to make sure all are numeric values
ccm_stats_num.dtypes
```

```python
# Get correlations and heat map
ccm_stats_num_correlations = ccm_stats_num.corr()
plt.figure(figsize=(20,11))
sns.heatmap(ccm_stats_num_correlations, annot = True, cmap = "coolwarm");
```
